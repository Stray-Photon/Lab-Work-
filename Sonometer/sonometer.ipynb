```python
import numpy as np
from matplotlib import pyplot as plt  
from scipy.stats import linregress
```

```python
g = 9.8
u = 0.31  #mass per unit length for the nylon string

M = np.array([1000, 1500, 2000, 2500])
```

```python
# for freq 256Hz
l1 = np.array([41, 45, 51, 56])

slope, intercept, r_value, p_value, std_err = linregress(np.sqrt(M), l1)


plt.plot(np.sqrt(M), l1, 'bo', label='Original Data')
plt.plot(np.sqrt(M), slope * np.sqrt(M) + intercept, 'r-', label='Fitted Line')
plt.grid()
plt.xlabel('$\sqrt{M}$')
plt.ylabel('Resonant length (L)')
plt.title('$\sqrt{M}$ vs $L$ for 256Hz')
plt.legend()
plt.savefig('256Hz.png') 

print("The slope of the fitted line is:", slope)
```

```python
def calculate_n(f, L, M, g, u):
    n = (2 * L * f * np.sqrt(u / (M * g)))/100
    return n

# Input values
f_value = 256  
L_value = np.array([41,45,51,56])  


result_n = calculate_n(f_value, L_value, M, g, u)
print("The value of n is:", result_n)
```
