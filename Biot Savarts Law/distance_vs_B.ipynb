```python
d1,b1 = np.loadtxt('dist_vs_B_i=5.dat',delimiter=',',unpack=True)
d2,b2 = np.loadtxt('dist_vs_B_i=10.dat',delimiter=',',unpack=True)
d3,b3 = np.loadtxt('dist_vs_B_i=15.dat',delimiter=',',unpack=True)
```
