```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
i1,B1 = np.loadtxt('i_vs_B_for,r=40.dat',delimiter=',',unpack=True)
i2,B2 = np.loadtxt('i_vs_B_for,r=80.dat',delimiter=',',unpack=True)
i3,B3 = np.loadtxt('i_vs_B_for,r=120.dat',delimiter=',',unpack=True)
```

```python
plt.scatter(i1,B1,color='black')
plt.scatter(i2,B2,color='black')
plt.scatter(i3,B3,color='black')
plt.plot(i1,B1,'black',label='for r=40mm')
plt.plot(i2,B2,'r',label='for r=80mm')
plt.plot(i3,B3,'y',label='for r=120mm')
plt.title('Current vs Magnetic Field')
plt.xlabel('Magnetic Field (in Gauss)')
plt.ylabel('Current (in A)')
plt.grid()
plt.legend(loc='lower right')
```
