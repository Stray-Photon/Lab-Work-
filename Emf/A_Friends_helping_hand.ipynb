```python
import matplotlib.pyplot as plt
import math as m

x=[60,50,45,40,30]
y1=[2.25,2,1.8,1.6,1.35]
y2=[2.23,1.95,1.80,1.60,1.20]
y3=[2.20,1.94,1.65,1.45,1.10]

s1=0
s2=0
s3=0
sx=0
for i in range(len(x)):
  xy1=x[i]*y1[i]
  s1+=xy1
  xy2=x[i]*y2[i]
  s2+=xy2
  xy3=x[i]*y3[i]
  s3+=xy3
  xx=x[i]*x[i]
  sx+=xx

m1=s1/sx
m2=s2/sx
m3=s3/sx

yl1=[]
yl2=[]
yl3=[]

xl=range(0,61)

for j in xl:
  yl1.append(m1*j)
  yl2.append(m2*j)
  yl3.append(m3*j)

plt.plot(x,y1,'o',color='blue')
plt.plot(xl,yl1,'b')

plt.plot(x,y2,'o',color='red')
plt.plot(xl,yl2,'r')

plt.plot(x,y3,'o',color='yellow')
plt.plot(xl,yl3,'y')

plt.xlabel("Deflection Angle (θ_o)")
plt.ylabel("Peak Voltage (ε) (V)")

plt.title("Constant Time Period\nl=25cm: Blue, l=20cm: Red, l=16cm: Yellow\nl = Distance of Each Mass from the Centre\n")

plt.grid()
plt.show()
```
