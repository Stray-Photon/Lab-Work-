```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
##for making a graph of peakvoltage vs angle

a1,T1,V1=np.loadtxt('data,l=10.dat',delimiter=',',unpack=True)
a2,T2,V2=np.loadtxt('data,l=16.dat',delimiter=',',unpack=True)
a3,T3,V3=np.loadtxt('data,l=20.dat',delimiter=',',unpack=True)
a4,T4,V4=np.loadtxt('data,l=25.dat',delimiter=',',unpack=True)
```

```python
def f(x,a,b):
    return a*x+b
popt,pcov=curve_fit(f,V1,a1)
x1Fit=np.arange(0,3,0.1)

def g(x,a,b):
    return a*x+b
popt,pcov=curve_fit(g,V2,a2)
x2Fit=np.arange(0,3,0.01)

def h(x,a,b):
    return a*x+b
popt,pcov=curve_fit(h,V3,a3)
x3Fit=np.arange(0,3,0.01)

def i(x,a,b):
    return a*x+b
popt,pcov=curve_fit(i,V4,a4)
x4Fit=np.arange(0,3,0.01)
```

```python
#plotting the points

plt.plot(V1,a1,'bo',ms=4,label='For l=10')
plt.plot(V2,a2,'go',ms=4,label='For l=16')
plt.plot(V3,a3,'ro',ms=4,label='For l=20')
plt.plot(V4,a4,'ko',ms=4,label='For l=25')

#curve fitting (ive taken the best fit I was getting among there four sets of data)

#plt.plot(x1Fit,f(x1Fit,*popt),'b--',label='Data fiiting for l=10')
#plt.plot(x2Fit,g(x2Fit,*popt),'g--',label='Data fiiting for l=16')
plt.plot(x3Fit,h(x3Fit,*popt),'m--',label='Data fiiting')
#plt.plot(x4Fit,i(x4Fit,*popt),'k--',label='Data fiiting for l=25')


plt.xticks(np.arange(0,3,0.5))
plt.yticks(np.arange(0,70,10))
plt.legend(loc='lower right')
plt.ylabel('Angle of deflection (in $theta$)')
plt.xlabel('Peak Voltage (in $v_o$)')
plt.title('Voltage vs Angle')
plt.show()
```
