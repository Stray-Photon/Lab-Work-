```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def C(f,R,Vr):
    Vo=10.8
    return (Vr/(2*np.pi*f*R*np.sqrt(Vo**2 - Vr**2)))*10**12

dist = [2,4,6] #in mm
dist_inv = []  #in mm^-1
for j in dist:
    dist_inv.append(1/j)

Vr = [0.031,0.026,0.024]
Capa = []
for i in Vr:
    Capa.append(C(1003,0.49*10**(6),i))

print(Capa)
```

```python
def f(x,a,b):
    return a*x+b
popt,pcov=curve_fit(f,dist_inv,Capa)
xFit=np.arange(0,0.6,0.01)
```

```python
plt.title('Variation of Capacitance with Distance between the plates')
plt.plot(dist_inv,Capa,'bo',ms=5,label='Experimental Data')
plt.plot(xFit,f(xFit,*popt),'--r',label='Data Fitting')
plt.legend(loc='lower right')
plt.yticks(np.arange(0,1,0.25))
plt.xlabel('1/Distance between the plates (in $mm^-1$)')
plt.ylabel('Capacitance(in pF)')
plt.grid()
plt.show
```
