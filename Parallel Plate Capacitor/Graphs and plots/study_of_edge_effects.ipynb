```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#we are supposed to make two graphs in this section
#one of Cxd(in pF.mm) vs d(in  mm)
#and of C/Area(in pF/cm^2) vs 1/Area(in cm^2)

dist = [2,4,6] #in mm
Capa1 = [0.9295287909591474, 0.7796038399619768, 0.7196340053242493]
Cxd = []

for i,j in zip(dist,Capa1):
    Cxd.append(i*j)

print(Cxd)
```

```python
def f(x,a,b):
    return a*x+b
popt,pcov=curve_fit(f,dist,Cxd)
xFit=np.arange(0,6.5,0.1)
```

```python
plt.title('Study of edge effects - graph 1')
plt.plot(dist,Cxd,'bo',ms=5,label='Experimental Data')
plt.plot(xFit,f(xFit,*popt),'--r',label='Data Fitting')
plt.yticks(np.arange(0,5.5,0.5))
plt.legend(loc='lower right')
plt.grid()
plt.xlabel('Distance (in mm)')
plt.ylabel('Capacitance x Distance (in pF.mm)')
```

```python
#this one is the second graph for edge effects

Area = [225,305.45,918.09] #in cm**2
invArea = []

for i in Area:
    invArea.append(1/i)

Capa2 = [0.8095887869728767, 0.8995437560286768, 1.5892069740934613]
CapainvArea = []

for a,b in zip(invArea, Capa2):
    CapainvArea.append(a*b)

print(CapainvArea)
```

```python
def f(x,a,b):
    return a*x+b
popt,pcov=curve_fit(f,invArea,CapainvArea)
xFit=np.arange(0,0.01,0.000001)
```

```python
plt.title('Study of edge effects graph 2')
plt.plot(invArea, CapainvArea,'bo',ms=5,label='Experimental Data')
plt.plot(xFit,f(xFit,*popt),'r--',label='Data fitting')
plt.yticks(np.arange(0,0.01,0.0005))
plt.legend(loc='lower right')
plt.grid()
plt.xlabel('1/Area (in $cm^{-2}$)')
plt.ylabel('Capacitance/Area (in pF/$cm^{-2}$)')
```
