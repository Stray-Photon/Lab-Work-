```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def C(f,R,Vr):
    Vo=10.8
    return (Vr/(2*np.pi*f*R*np.sqrt(Vo**2 - Vr**2)))*10**12

Area = [225,305.45,918.09] #in cm**2
Vr = [0.027,0.030,0.053]
Capa = []
for i in Vr:
    Capa.append(C(1003,0.49*10**(6),i))

print(Capa)
```

```python
def f(x,a,b):
    return a*x+b
popt,pcov=curve_fit(f,Area,Capa)
xFit=np.arange(0,1000,0.1)
```

```python
plt.plot(xFit,f(xFit,*popt),'r--',label='Data fiiting')
plt.plot(Area, Capa,'bo',ms=5,label='Experimental data')
plt.grid()
plt.yticks(np.arange(0,2,0.25))
plt.xlabel('Area of the plates(in $cm^2$)')
plt.ylabel('Capacitance(in pF)')
plt.legend(loc='lower right')
plt.title('Variation of Capacitance with the Change of Area of the plates')
plt.show()
```
